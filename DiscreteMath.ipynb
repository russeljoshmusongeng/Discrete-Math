```python
# Odd-Indexed Insertion Sort (GENERAL VERSION)
# Gumagana sa kahit anong input na ibigay ng user
# Only elements at ODD INDICES (1,3,5,...) ang keys
# Even indices hindi ginagalaw unless nadamay sa shift

# Halimbawa ng input:
# 1.1 5.5 2.2 4.4 3.3
A = list(map(float, input("Enter the list of numbers separated by space: ").split()))

n = len(A)        # haba ng list
shifts = 0        # bilang ng total shifts

# Loop sa odd indices (Python is 0-based)
for i in range(1, n, 2):
    key = A[i]        # element na i-iinsert
    j = i - 2         # previous odd index

    # While mas malaki ang previous odd element kaysa key
    while j >= 0 and A[j] > key:
        A[j + 2] = A[j]     # shift ng element two positions to the right
        j = j - 2           # move to next previous odd index
        shifts += 1         # dagdag shift counter

    # Insert ng key sa tamang odd position
    A[j + 2] = key

# Output
print("\nFinal Sorted List:", A)
print("Total Number of Shifts:", shifts)
```
